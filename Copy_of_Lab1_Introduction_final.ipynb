```python
# C. look for missing values
data.isnull().sum()
data.isna().sum() / data.shape[0] * 100 # To find the percentage of the missing values

# If the column has just little missing values (Embarked), just get the rows with the NAN out. (Won't hurt the dataset much)
# If the column has too many missing values, just GET the whole column OUT (Best Practice)

```

As around 77% of `Cabin` are missing, so it's best to remove the entire column.

```python
# Drop rows
data_cleaned = data.drop(columns = ['Cabin']) # inplace to make changes to the actual data (without creating a new DataFrame or assign to a new variable)
data_cleaned.shape

# data_cleaned.dropna(subset=['Cabin']) # remove rows with at least one missing value.
  # subset: to specify the column you want to dropna to!
```

```python
# Drop rows with at leasrt 1 missing value:
data_cleaned.dropna(inplace=True)
data_cleaned.shape
```

```python
# Check again, if there is any more null
data_cleaned.isnull().sum() / len(data) * 100
```

```python
# F.How many passengers were younger than 3 years old? How many were older than 60 years old?\
data_cleaned['Age']

under_3 = data_cleaned[data_cleaned['Age'] < 3]
over_60 = data_cleaned[data_cleaned['Age'] > 60]

print(f"1. Under 3: {len(under_3)}\n2. Over 60: {len(over_60)}")
```

```python
# orrrr, using .query
under_3_q = data_cleaned.query("Age < 3").shape[0]
over_60_q = data_cleaned.query("Age > 60").shape[0]

print(f"1. Under 3: {under_3_q}\n2. Over 60: {over_60_q}")
```

### **2.3. Multiple information**

**I.** How many 1st class passengers survived? How about 2nd and 3rd class?

**J.** How many female passengers survived? How many males did?

**K.** How many people from each embarkation port survived?

**L.** Was `Jack` on the ship? How about `Rose`?

```python
# To do
# I. Survived passengers from each of seat classes
survived_data = data_cleaned.copy()
survived_data['Survived'] = survived_data['Survived'].replace({0: 'Not Survied', 1: 'Survived'})
survived_data['Pclass'] = survived_data['Pclass'].replace({
    1: '1st Class',
    2: '2nd Class',
    3: '3rd Class'
})
    # Replacing value names, js for fun :)

survived_data.groupby(['Survived', 'Pclass']).size()
```

```python
# J. Suvival Count by Gender
data_cleaned.groupby(['Survived', 'Sex']).size()
```

```python
# K. Survival Count by Port of Embarkment
data_cleaned.groupby(['Survived', 'Embarked']).size()
```
