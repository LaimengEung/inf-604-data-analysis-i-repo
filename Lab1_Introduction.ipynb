### **2.2. Single information**


**D.** How many male and female passengers were on the ship?

**E.** How many of them survived? How many didn't?

**F.** How many passengers were younger than 3 years old? How many were older than 60 years old?

**G.** How many passengers embarked from the three ports?

- `C`: Cherbourg, France.
- `Q`: Queentown, Ireland.
- `S`: Southampton, England.

**H.** How many passengers were in the 1st, 2nd and 3rd class?


```python
# To do
# D. Male and Female passengers aboard
data_cleaned['Sex']

male = len(data_cleaned[data_cleaned['Sex'] == 'male'])
female = len(data_cleaned[data_cleaned['Sex'] == 'female'])

print(f"There are {male} male, and {female} female")

# ORRRRRRR,
m = data_cleaned['Sex'].value_counts()[0]

data_cleaned['Sex'].value_counts(normalize=True) # To show proportion (in percentage)
```

```python
# E. How many of them Survived?
data_cleaned['Survived']

alive = len(data_cleaned[data_cleaned['Survived'] == 1])
dead = len(data_cleaned[data_cleaned['Survived'] == 0])

print(f"There are {alive} alive, and {dead} dead people")
```

```python
# F.How many passengers were younger than 3 years old? How many were older than 60 years old?\
data_cleaned['Age']

under_3 = data_cleaned[data_cleaned['Age'] < 3]
over_60 = data_cleaned[data_cleaned['Age'] > 60]

print(f"1. Under 3: {len(under_3)}\n2. Over 60: {len(over_60)}")
```

```python
# ORRRRRR, USING .query
over_60_query = data_cleaned.query("Age > 60").shape
over_60_query[0]
```
